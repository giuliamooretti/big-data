1. Crea il DataFrame **vendite** che contiene la quantità di torte, biscotti e taralli (colonne) nei diversi giorni della settimana (righe).

  Inserisci valori casuali compresi tra 5 e 50 (np.random.seed(100)).

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(100)

n_week_days = 7
week_days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
data = {
    "cakes" : np.random.randint(5, 50, size=7),
    "biscuits" : np.random.randint(5, 50, size=7),
    "taralli" : np.random.randint(5, 50, size=7) 
}

sells = pd.DataFrame(data, index=week_days)
sells
```

2. Visualizza le vendite di torte durante tutta la settimana.

```python
sells['cakes']
```

3. Ottieni le vendite del giovedi.

```python
sells.loc['Thursday']
```

4. Visualizza le vendite di biscotti e taralli dal mercoledì alla domenica.

```python
sells[['biscuits', 'taralli']].loc['Wednesday':]
```

5. Qual è il totale delle vendite per ogni prodotto durante la settimana?

```python
sells.sum()
```

6. Qual è il giorno con il massimo delle vendite di torte?

```python
sells['cakes'].idxmax()
```

7. Aggiungi la colonna **Totale Vendite Giorno** che rappresenta la somma delle vendite di tutti i prodotti per ciascun giorno.

```python
sells.reindex(columns=['cakes', 'biscuits', 'taralli', 'total'], fill_value=0)
sells['total'] = sells[['cakes', 'biscuits', 'taralli']].sum(axis=1)
sells
```

8. Quanti giorni hanno avuto un totale di vendite giornaliere superiori a 80?

```python
(sells['total'] > 80).sum()
```
