# Transform

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

## Load Dataset from CSV

```python
dataset = pd.read_csv('./dataset.csv')
dataset.head()
```

## Convert Column Data Type

```python
df = dataset.copy()
df['MonthlyCharges'] = df['MonthlyCharges'].astype('object')
df['MonthlyCharges'].dtype
```

## Binary Features

```python
print(dataset['gender'].unique())
df = dataset.copy()
df['gender'] = df['gender'].map({'Female' : 1, 'Male' : 0})
df['gender'].head()
```

Multiple automatic binarizations

```python
cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
for col in cols: 
    print(f'{col}: {dataset[col].unique()}')
```

```python
df = dataset.copy()
for col in cols:
    df[col] = df[col].map({'No' : 0, 'Yes' : 1, 'No internet service' : 0})
df.head()
```

## Binning

```python
pd.cut(dataset['tenure'], 3, labels=['low', 'average', 'high'])
```

## Categorical Features

### To Label

```python
cat_features = ['InternetService', 'Contract', 'PaymentMethod']
df = dataset.copy()
for cat_feature in cat_features:
    df[cat_feature] = LabelEncoder().fit_transform(df[cat_feature])
df.head()
```

### To Dummy

```python
cat_features = ['InternetService', 'Contract', 'PaymentMethod']
df = dataset.copy()
for cat_feature in cat_features:
    df = pd.concat([df, pd.get_dummies(df[cat_feature])], axis=1).drop(columns=cat_feature)
df.head()
```
