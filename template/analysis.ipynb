## GroupBy

```python
print(dataset.groupby('InternetService')['tenure'].groups)
dataset.groupby('InternetService')['tenure'].mean()
```

## Pivot Table

```python
dataset.pivot_table(index=['InternetService', 'Contract'], values=['Churn'], aggfunc='count')
```

### Count-to-Percentage

Global

```python
dataset['Churn'] = dataset['Churn'].map({'Yes' : 1, 'No' : 0})
pivot = dataset.pivot_table(index=['InternetService'], columns=['Contract'], values=['Churn'], aggfunc='mean', observed=True)
pivot
```

Local

```python
pivot = dataset.pivot_table(index=['InternetService', 'Contract'], values=['Churn'], aggfunc='count')
pivot['Churn'] / pivot.groupby('InternetService')['Churn'].sum()
```
