2. Crea una nuova colonna chiamata **Livello Esperienza**, utilizzando la colonna **Anni di esperienza**. Se l'esperienza è minore di 5 anni, mappa il valore a "Junior", tra 5 e 10 a "Mid" e più di 10 a "Senior".

```python
def find_level(years):
    if years < 5:
        return 'Junior'
    elif years < 10:
        return 'Mid'
    else:
        return 'Senior'
    
# data.insert(len(data.columns),'Livello Esperienza', data['Anni di esperienza'].map(find_level), True)
data['Livello Esperienza'] = data['Anni di esperienza'].map(find_level)
data
```

3. Sostituisci i nomi dei dipartimenti con le seguenti abbreviazioni: Vendite -> 'VEN', Information Technology -> 'IT', Human Resources -> 'HR'. Quanti dipendenti ci sono per ogni dipartimento?

```python
dip_acronyms = {
    'Vendite' : 'VEN',
    'Information Technology' : 'IT',
    'Human Resources' : 'HR'
}

data['Dipartimento'] = data['Dipartimento'].map(dip_acronyms)
data
```
